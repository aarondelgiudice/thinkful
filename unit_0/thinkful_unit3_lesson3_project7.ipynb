```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# create two populations and samples using numpy's binomial function
# the populations should have n equal to 10 and size equal to 10000.
# The p-value of pop1 should be 0.2 and the p-value of pop2 should be 0.5.
# Using a sample size of 100, calculate the means and standard deviations of your samples

pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10, 0.5, 10000)

pop1_mu = pop1.mean()
pop1_std = pop1.std()

pop2_mu = pop2.mean()
pop2_std = pop2.std()

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

sample1_mu = sample1.mean()
sample1_std = sample1.std()

sample2_mu = sample2.mean()
sample2_std = sample2.std()

print("population 1 mean:" , pop1_mu)
print("population 1 std:" , pop1_std)
print("population 2 mean:" , pop2_mu)
print("population 2 std:" , pop2_std)
print("sample 1 mean:" , sample1_mu)
print("sample 1 std:" , sample1_std)
print("sample 2 mean:" , sample2_mu)
print("sample 2 std:" , sample2_std)
```

```python
plt.figure(figsize=(6,4))

plt.subplot(1, 2, 1)
plt.hist(sample1, alpha=.6, label='sample 1')
plt.hist(sample2, alpha=.6, label="sample 2")
plt.legend(loc='upper right')

plt.subplot(1, 2, 2)
plt.hist(pop1, alpha=.6, label='population 1')
plt.hist(pop2, alpha=.6, label="population 2")
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

```python
# Increase the size of your samples from 100 to 1000
# calculate the means and standard deviations for your new samples and create histograms for each

sample1_1000 = np.random.choice(pop1, 1000, replace=True)
sample2_1000 = np.random.choice(pop2, 1000, replace=True)

sample1_1000_mu = sample1_1000.mean()
sample1_1000_std = sample1_1000.std()

sample2_1000_mu = sample2_1000.mean()
sample2_1000_std = sample2_1000.std()

plt.figure(figsize=(6,4))

plt.subplot(1, 2, 1)
plt.hist(sample1_1000, alpha=.6, label='sample 1')
plt.hist(sample2_1000, alpha=.6, label="sample 2")
plt.legend(loc='upper right')

plt.subplot(1, 2, 2)
plt.hist(pop1, alpha=.6, label='population 1')
plt.hist(pop2, alpha=.6, label="population 2")
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()

print("for n=1000, sample1 mean: {}, sample2 mean: {}, sample1 std: {}, sample2 std: {}"
      .format(sample1_1000_mu, sample2_1000_mu, sample1_1000_std, sample2_1000_std))
print("for n=100, sample1 mean: {}, sample2 mean: {}, sample1 std: {}, sample2 std: {}"
      .format(sample1_mu, sample2_mu, sample1_std, sample2_std))
print("pop1 mean: {}, pop2 mean: {}, pop1 std: {}, pop2 std: {}".format(pop1_mu, pop2_mu, pop1_std, pop2_std))

print("As the sample size n increases, we can see that the sample data more closely matches the population data.")
```

```python
# Decrease the size of your samples to 20
# calculate the means and standard deviations for your new samples and create histograms for each

sample1_20 = np.random.choice(pop1, 20, replace=True)
sample2_20 = np.random.choice(pop2, 20, replace=True)

sample1_20_mu = sample1_20.mean()
sample1_20_std = sample1_20.std()

sample2_20_mu = sample2_20.mean()
sample2_20_std = sample2_20.std()

plt.figure(figsize=(6,4))

plt.subplot(1, 2, 1)
plt.hist(sample1_20, alpha=.6, label='sample 1')
plt.hist(sample2_20, alpha=.6, label="sample 2")
plt.legend(loc='upper right')

plt.subplot(1, 2, 2)
plt.hist(pop1, alpha=.6, label='population 1')
plt.hist(pop2, alpha=.6, label="population 2")
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()

print("for n=20, sample1 mean: {}, sample2 mean: {}, sample1 std: {}, sample2 std: {}"
      .format(sample1_20_mu, sample2_20_mu, sample1_20_std, sample2_20_std))
print("for n=100, sample1 mean: {}, sample2 mean: {}, sample1 std: {}, sample2 std: {}"
      .format(sample1_mu, sample2_mu, sample1_std, sample2_std))
print("pop1 mean: {}, pop2 mean: {}, pop1 std: {}, pop2 std: {}".format(pop1_mu, pop2_mu, pop1_std, pop2_std))

print("As the sample size n decreases, the sample data matches the population data less")
```

```python
# Change the probability value (p in the NumPy documentation) for pop1 to 0.3
# take new samples and compute the t-statistic and p-value
# change the probability value p for group 1 to 0.4, and do it again
# What changes, and why?

print("The smaller the p-value, the more likely the sample data matches the population data")
print("Increasing the p-value of pop1 will decrease the chance that the sample data will match the population data")
print("Decreasing the p-value of pop2 will increase the chance that the sample data will match the population data")
```

```python
# Change the probability value (p in the NumPy documentation) for pop1 to 0.3
# take new samples and compute the t-statistic and p-value

pop1_new = np.random.binomial(10, 0.3, 10000)

pop1_new_mu = pop1_new.mean()
pop1_new_std = pop1_new.std()

sample1_new = np.random.choice(pop1_new, 100, replace=True)

sample1_new_mu = sample1_new.mean()
sample1_new_std = sample1_new.std()

diff = sample1_new_mu - sample1_mu

print("population 1 mean:" , pop1_mu)
print("population 1 std:" , pop1_std)
print("new population 1 mean:" , pop1_new_mu)
print("new population 1 std:" , pop1_new_std)
print("new sample 1 mean:" , sample1_new_mu)
print("new sample 1 std:" , sample1_new_std)
print("difference between sample means: ", diff)

from scipy.stats import ttest_ind
print(ttest_ind(sample1_new, sample1, equal_var=False))
```
