```python
import numpy as np
import pandas as pd
import statistics as stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Choose 6 distributions and generate a random variable with 100 datapoints
# Graph the variable using a histogram

# Compute the mean and standard deviation and plot them as vertical lines on the histogram

# Evaluate whether the descriptive statistics provided useful information about the variable.
# Can you identify any common characteristics of the distributions
# any that could be usefully described using the mean and/or standard deviation, versus the ones that could not?
```

```python
# Additionally Generate two normally-distributed variables

# one with a mean of 5 and standard deviation of 0.5
val1 = np.random.normal(5, 0.5, 100)

# the other with a mean of 10 and standard deviation of 1
val2 = np.random.normal(10, 1, 100)

# Add them together to create a third variable and graph the third variable using a histogram
val3 = val1 + val2

# Compute the mean and standard deviation and plot them as vertical lines on the histogram.
# Evaluate the descriptive statistics against the data.

plt.hist(val3)
plt.title("Normal Distribution")

# Add a vertical line at the mean.
plt.axvline(val3.mean(), color='y', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(val3.mean() + val3.std(), color='y', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(val3.mean()-val3.std(), color='y', linestyle='dashed', linewidth=2)

plt.show()
```

```python
beta = np.random.beta(.4, .6, 100)

plt.figure(figsize=(6, 4))

plt.subplot(1, 2, 1)
plt.hist(beta)
plt.title("Beta Distribution")

# Add a vertical line at the mean.
plt.axvline(beta.mean(), color='y', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(beta.mean() + beta.std(), color='y', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(beta.mean()-beta.std(), color='y', linestyle='dashed', linewidth=2)

#
#sort values in ascending order
val3.sort()
beta.sort()

# Plot the variable rand1 against norm in qqplots.
plt.subplot(1, 2, 2)
plt.plot(val3, beta, 'o')
plt.title("QQ Plot")

plt.tight_layout()
plt.show()

print("Beta distribution is a continuous probability distribution.")
print("The summary stats are a mixed bag. The mean seems to accurately bisect the data.")
print("But the data is polar, clustered around the min and max values,")
print("and the standard deviation doesn't offer any significant insight.")
```

```python
expo = np.random.exponential(10000, 100)

plt.figure(figsize=(6, 4))

plt.subplot(1, 2, 1)
plt.hist(expo)
plt.title("Exponential Distribution")

# Add a vertical line at the mean.
plt.axvline(expo.mean(), color='y', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(expo.mean() + expo.std(), color='y', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(expo.mean()-expo.std(), color='y', linestyle='dashed', linewidth=2) 

#
#sort values in ascending order
val3.sort()
expo.sort()

# Plot the variable rand1 against norm in qqplots.
plt.subplot(1, 2, 2)
plt.plot(val3, expo, 'o')
plt.title("QQ Plot")

plt.tight_layout()
plt.show()

print("Exponential distribution describes the time between events that occur independently at a constant rate.")
print("The majority of the data falls within one standard deviation of the mean,")
print("even though the data is mostly clustered at the lower end of the scale.")
print("the QQ plot gives use a better idea of how far the data deviates from a normal distribution.")
print("We can see some outliers at the higher end and how the data points start to spread out.")
```

```python
log = np.random.lognormal(5, .25, 100)

plt.figure(figsize=(6, 4))

plt.subplot(1, 2, 1)
plt.hist(log)
plt.title("Lognormal Distribution")

# Add a vertical line at the mean.
plt.axvline(log.mean(), color='y', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(log.mean() + log.std(), color='y', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(log.mean()-log.std(), color='y', linestyle='dashed', linewidth=2) 

# 
#sort values in ascending order
val3.sort()
log.sort()

# Plot the variable rand1 against norm in qqplots.
plt.subplot(1, 2, 2)
plt.plot(val3, log, 'o')
plt.title("QQ Plot")

plt.tight_layout()
plt.show()

print("Log-normal distribution is a continuous distribution of a random variable whose logarithm is normally distributed.")
print("As such, the data closely follows a normal distribution.")
print("The summary stats are indicative of the nature of the data and the QQ plot shows a normal distribution.")
```

```python
pareto = np.random.pareto(5, 100)

plt.figure(figsize=(6, 4))

plt.subplot(1, 2, 1)
plt.hist(pareto)
plt.title("Pareto Distribution")

# Add a vertical line at the mean.
plt.axvline(pareto.mean(), color='y', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(pareto.mean() + pareto.std(), color='y', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(pareto.mean()-pareto.std(), color='y', linestyle='dashed', linewidth=2)

# 
#sort values in ascending order
val3.sort()
pareto.sort()

# Plot the variable rand1 against norm in qqplots.
plt.subplot(1, 2, 2)
plt.plot(val3, pareto, 'o')
plt.title("QQ Plot")

plt.tight_layout()
plt.show()

print("Pareto distribution is a power-law probability distribution.")
print("Like a poisson distribution, the values of a pareto distribution are largely clustered at one end of the scale.")
print("As such, the descriptive statistic are not very insightful.")
print("The QQ plot confirms the data does not follow a normal distribution")
```

```python
ray = np.random.rayleigh(.5, 100)

plt.figure(figsize=(6, 4))

plt.subplot(1, 2, 1)
plt.hist(ray)
plt.title("Rayleigh Distribution")

# Add a vertical line at the mean.
plt.axvline(ray.mean(), color='y', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(ray.mean() + ray.std(), color='y', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(ray.mean()-ray.std(), color='y', linestyle='dashed', linewidth=2) 

# 
#sort values in ascending order
val3.sort()
ray.sort()

# Plot the variable rand1 against norm in qqplots.
plt.subplot(1, 2, 2)
plt.plot(val3, ray, 'o')
plt.title("QQ Plot")

plt.tight_layout()
plt.show()

print("The rayleigh distribution is a chi distribution of positive-valued random variables in two degrees of freedom."+
     "\nThe data is largely contained within one standard deviation of the mean."+
     "\nOutside of a few outliers at the higher end of the scale, the data behaves largely like a normal distribution.")
```

```python
wald = np.random.wald(4, 5, 100)

plt.figure(figsize=(6, 4))

plt.subplot(1, 2, 1)
plt.hist(wald)
plt.title("Wald Distribution")

# Add a vertical line at the mean.
plt.axvline(wald.mean(), color='y', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(wald.mean() + wald.std(), color='y', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(wald.mean()-wald.std(), color='y', linestyle='dashed', linewidth=2)

# 
#sort values in ascending order
val3.sort()
wald.sort()

# Plot the variable rand1 against norm in qqplots.
plt.subplot(1, 2, 2)
plt.plot(val3, wald, 'o')
plt.title("QQ Plot")

plt.tight_layout()
plt.show()

print("The wald distribution, or inverse guassian distribution, is a two parameter distribution with support."+
     "\nThe descriptive stats aren't very informative and the QQ plot shows how far the data varys from a normal distribution.")
```
